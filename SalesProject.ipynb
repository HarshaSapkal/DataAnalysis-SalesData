# Cleaning the order date and shipdate columns

```python
df['Order Date']=pd.to_datetime(df['Order Date'])  
df
#The dataset was previously having random or not same datetime format
#therefore we changed the whole datetime format to (YYYY-MM-DD)
```

```python
df['Ship Date']=pd.to_datetime(df['Ship Date'])
df
```

# Checking the overall sales trend

```python
df['Order Date'].min()
```

```python
df['Order Date'].max()
```

```python
# Getting only month and year column from Order Date
```

```python
df['month_year']=df['Order Date'].apply(lambda x: x.strftime('%Y-%m'))
df['month_year']
```

```python
df_trend=df.groupby('month_year').sum()['Total Profit'].reset_index()
df_trend
```

```python
#plotting graph year vs profit
```

```python
plt.figure(figsize=(18,6))
plt.plot(df_trend['month_year'],df_trend['Total Profit'])
plt.xticks(rotation='vertical',size=8)
plt.show()
#Company is being consistant over years.No major growth observed.
#2013-01 2013-02 comapny gained high profit above 0.9.
#2013-05 company had very low profit.
#2013-08 to 2015-04 company had consitant profit rate.
#2015-05 company had very less profit less that 0.1.
#then in 2016 end and 2017 start they gained maajor profit.
```

```python
mprofit=df['Total Profit'].mean()
meprofit=df['Total Profit'].median()

sns.distplot(df['Total Profit'])
plt.axvline(mprofit,color='r')
plt.axvline(meprofit,color='y')
plt.grid(True)
plt.show()
```

# Which country has the highest purchase count

```python
plt.figure(figsize=(12,30))
sns.countplot(data=df,y='Country')
plt.tight_layout()
plt.show()
# cuba country is the highest followed by madagaskar,zimbabwe,bahrain.
#cameroon,malta,kosovo is the least.
```

# Profit according to products

```python
plt.figure(figsize=(15,6))
sns.lineplot(df['Item Type'],df['Total Profit'])
plt.grid(True)
plt.show()
#Cosmetics has the highest profit earned followed by Household,suppliers and baby food.
#while fruits has the lowest followed by beverages ,personal care and meat in increasing order of profit.
```

# Fetching top products by its units sold

```python
#Grouping item type & units sold
```

```python
Item_Sold=df.groupby('Item Type').sum()['Units Sold'].reset_index()
Item_Sold.sort_values('Units Sold',ascending=False)
#All the items arranged according to their units sold.Beverages has most units sold.
```

# Plotting graph- units sold of all products

```python
plt.figure(figsize=(10,5))
plt.plot(Item_Sold['Item Type'],Item_Sold['Units Sold'])
plt.xticks(rotation='vertical',size=8)
plt.show()
# From the graph we can see that Beverages has the highest units sold and fruits has the lowest units sold.
# vegetables and personal care-almost same amount of units sold although personal care has more.
```

# Fetching Item type according to their Total cost

```python
plt.figure(figsize=(15,6))
sns.lineplot(df['Item Type'],df['Total Cost'])
plt.grid(True)
plt.show()
#Office suppliers has the highest costing.followed by household and meat.
#Fruit has the lowest costing.
```

# Item type and their total unit cost

```python
UnitCost=df.groupby('Item Type').sum()['Unit Cost'].reset_index()
UnitCost.sort_values('Unit Cost',ascending=False)
#all the item type are sorted according their descending order of unit cost.
#Office suppliers has the highest unit cost followed by household and meat.Fruits has the lowest unit cost.
```
